```python
df[["WHO Region","Continent","Country/Region"]].tail(30)
```

```python
def country(a):
    if a=="North America" or a=="	South America":
        return "Americas"
    elif a=="Asia":
        return "South-EastAsia"
    elif a=="Europe":
        return "Europe"
    elif a=="Australia/Oceania":
        return "WesternPacific"   
    elif a=="Africa":
        return "Africa"     
```

```python
df["WHO Region"]=df["WHO Region"].fillna(df["Continent"].apply(country))
```
