```python
import pandas as pd
s1=pd.Series([i for i in range (11)], index=[i*3 for i in range(11)])
s1
```

```python
print(s1.mean())
print(s1.std())
```

```python
s2=pd.Series([i for i in range(5)], index=[i*10 for i in range(1,6)])
s2
```

```python
s3=pd.Series([2,4,6,8,10])
s4=pd.Series([1,3,5,7,9])
print(s3.add(s4))
print(s3.sub(s4))
print(s3.mul(s4))
print(s3.div(s4))
```
